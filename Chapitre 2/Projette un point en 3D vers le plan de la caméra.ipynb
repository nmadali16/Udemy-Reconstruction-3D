Pour convertir les coordonnées du monde en pixels, nous devons effectuer trois étapes :
- 1. Convertir le point mondial (X<sub>W</sub>,Y<sub>W</sub>,Z<sub>W</sub>) avec le centre optique **W</t>** à un nouveau système de coordonnées qui a le centre optique de la caméra **C</t>** comme origine (0,0,0). Nous
faites cela via une transformation de corps rigide, donc rotation R et translation t.
- 2. Convertissez le point de caméra résultant (X<sub>C</sub>,Y<sub>C</sub>,Z<sub>C</sub>) en nouvelles coordonnées dans le même système de coordonnées sur l'optique capteur, (x,y,z) en utilisant la matrice des paramètres intrinsèques des caméras **K**. Le capteur optique est également appelé « plan image » ou « cadre image ».
- 3. Convertissez (x,y,z) en coordonnées de pixels (u,v) en normalisant avec z et en décalant l'origine de l'image.

```python

keep_idxs=np.ones(num_images)
for i, fn in enumerate(img_names):
    


    img = cv2.cvtColor(cv2.imread(fn), cv2.COLOR_BGR2GRAY)

    
    found, corners = cv2.findChessboardCorners(img, pattern_size)  
    if not found:
        keep_idxs[i]=0
        print("chessboard not found")
        continue

    
    img_points.append(corners.reshape(-1, 2))
    obj_points.append(pattern_points)


img_names=[img_names[i] for i in range(num_images) if keep_idxs[i]==1]
```

## Du monde à la caméra

Étant donné les coordonnées du monde (X<sub>W</sub>, Y<sub>W</sub>, Z<sub>W</sub>), nous voulons trouver de nouvelles coordonnées (X<sub>C</sub> sub>, Y<sub>C</sub>, Z<sub>C</sub>) ca projection sur la caméra est donner par


$$\begin{bmatrix}
u\\ 
v\\ 
w
\end{bmatrix}=K[R \mid t]\begin{bmatrix}
X\\ 
Y\\ 
Z\\ 
1
\end{bmatrix}$$

```python

# Paramètres intrinsèques des caméras
K=camera_matrix
# Matrice de rotation en radian
R=cv2.Rodrigues(_rvecs[0])[0]
# Vecteur de translation en mètres
t=_tvecs[0].reshape(3,-1)
```

```python
img=cv2.imread(img_names[0])
plt.imshow(img)
plt.scatter(img_points[0][:,0],img_points[0][:,1])
```

Écrivons le code pour projeter l'origine (X, Y, Z) = (0, 0, 0) du système de coordonnées mondial dans les caméras
coordonnées des pixels (u, v).

Pour ce faire, créons d'abord une matrice T = [R|t], qui est juste une matrice 3x4 combinant la matrice de rotation R et
vecteur de traduction t.

```python
# Matrice de transformation T = [R|t]
T = np.concatenate((R, t), axis=1)
```

Pour transformer les coordonnées du monde (X<sub>W</sub>, Y<sub>W</sub>, Z<sub>W</sub>) vers le cadre de la caméra (X<sub>C</sub> , Y<sub>C</sub>, Z<sub>C</sub>), il suffit de multiplier *T* par
(X,Y,Z,1). Rappelez-vous que T est une matrice 3x4, donc pour ajouter le vecteur de traduction t, nous devons ajouter un
composant 1 au point de coordonnées du monde.

```python
# Origine dans le système de coordonnées mondial, située à (X,Y,Z,1) = (0, 0, 0, 1)
Pw = np.array(
    [[0],
     [0],
     [0],
     [1]]
)

# Projeter (Xw, Yw, Zw, 0) dans le système de coordonnées des caméras
Pc = np.matmul(T, Pw)

# Appliquer les intrinsèques de la caméra à la carte (Xc, Yc, Zc) à p=(x, y, z)
p = np.matmul(K, Pc)

# Normaliser par z pour obtenir (u,v,1)
uv = (p / p[2][0])[:-1]
uv
```

Ce que nous avons obtenu en retour, ce sont des coordonnées de pixels discrets (u,v) que nous pouvons tracer sur notre image.

```python
# Tracer l'origine du monde (0, 0, 0) sur les images aux coordonnées UV-Pixel normalisées
plt.imshow(img, cmap='gray')
plt.scatter(x=uv[0], y=uv[1], c='r', s=10)
plt.show()
```

Nous pouvons voir très clairement que l’origine du système de coordonnées mondiales est située dans le coin supérieur gauche de l’échiquier.
Soyons un peu plus sophistiqués et dessinons un cube en haut à droite de l'échiquier. En mesurant l’échiquier sur notre table, nous pouvons découvrir qu’un carré mesure 0,04 mx 0,04 m. Dans un premier temps, déplaçons donc notre point de 28 cm vers la droite. Pour ce faire, nous devons simplement redéfinir nos coordonnées mondiales pour qu'elles ne soient pas (X, Y, Z) = (0, 0, 0), mais (0,24, 0, 0).

Écrivons également deux fonctions rapides, l'une qui crée une matrice à partir de coordonnées x, y, z pour le monde réel, l'autre la transformant en coordonnées discrètes (u, v).

```python
def makeWorldVector(x=0, y=0, z=0):
    """
     Crée un vecteur de coordonnées mondiales à partir des composants x, y, z
     :numéro de paramètre x : coordonnée x du point en mètres
     :numéro de paramètre y : coordonnée y du point en mètres
     :param number z : coordonnée z du point en mètres
     :return : Vecteur en coordonnées mondiales
     :rtype: np.array(x, y, z, 1)
     """
    return np.array(
        [[x],
         [y],
         [z],
         [1]]
    )

def worldToPixels(Pw):
    """
     Convertit un vecteur en coordonnées mondiales en coordonnées de pixels UV
     :param np.array(x, y, z, 1) Pw : vecteur mondial
     :return : (u,v) coordonnées en pixels du point mondial dans le cadre de l'image
     :rtype : Tuple[nombre, nombre]
     """
    Pc = np.matmul(T, Pw)
    Pc_norml = (Pc / Pc[2][0])
    p = np.matmul(K, Pc_norml)
    uv = p[:-1]
    return uv[0], uv[1]

def plotPoints(us, vs):
    """
     Trace une liste de coordonnées de pixels sur une image
     :param List[number] us : Liste des coordonnées u pour les points du nuage de points
     :param List[number] vs : liste des coordonnées V pour les points du nuage de points
     :retour : Aucun
     """
    plt.imshow(img, cmap='gray')
    plt.scatter(x=us, y=vs, c='r', s=10)
    plt.show()

Pw = makeWorldVector(x=2, y=0)
u, v = worldToPixels(Pw)
plotPoints([u], [v])
```

Ensuite, définissons d'abord le cube en coordonnées mondiales, puis montrons-le en coordonnées UV en pixels.

```python
# Définir les coordonnées de départ dans le système de coordonnées mondiales (mètres) et la taille du pas dans chaque direction (mètres)
base_x, base_y, base_z = 0, 0, 0
delta_x, delta_y, delta_z = 6, 6, 6

# Créez des coins de cube en faisant une boucle sur les dimensions x, y, z et créez deux points pour chaque dimension
us, vs = [], []
for x in range(2):
    for y in range(2):
        for z in range(2):
            # Create a new point
            Pw = makeWorldVector(
                x=base_x+(delta_x*x), 
                y=base_y+(delta_y*y), 
                z=base_z+(delta_z*z)
            )
            # Convertissez-le en coordonnées de pixels UV et ajoutez-les au tableau us, vs
            u, v = worldToPixels(Pw)
            us.append(u)
            vs.append(v)

# Tracer les résultats
plotPoints([us], [vs])
```

Pour mieux illustrer le cube, codons en dur quelques lignes pleines

```python
img_plot = plt.imshow(img, cmap='gray')

def drawLine(from_n, to_n):
    """
     Trace une ligne sur le contexte de tracé actuel entre deux points donnés
     :param Tuple[int, int] from_n : (u,v) coordonnées du premier point
     :param Tuple[int, int] to_n : (u,v) coordonnées du deuxième point
     :retour : Aucun
     """
    plt.plot([us[from_n], us[to_n]], [vs[from_n], vs[to_n]], c='r')

drawLine(0, 1)
drawLine(1, 5)
drawLine(5, 4)
drawLine(4, 0)

drawLine(2, 3)
drawLine(3, 7)
drawLine(7, 6)
drawLine(6, 2)

drawLine(0, 2)
drawLine(1, 3)

drawLine(4, 6)
drawLine(5, 7)
    
plt.show()
```
