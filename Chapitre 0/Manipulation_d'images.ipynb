```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Lire  une image
image_path = "data/monalisa.jpg"
image = cv2.imread(image_path)
```

```python
# Afficher l'image en utilisant matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Image originale")
plt.axis("off")
plt.show()
```

```python
# Obtenir les propriétés de l'image
height, width, channels = image.shape
color_depth = image.dtype

print("Propriétés de l'image :")
print("Dimensions :", width, "x", height)
print("Nombre de canaux :", channels)
print("Profondeur de couleur :", color_depth)
```

```python
# Recadrer l'image
x, y, w, h = 100, 100, 32, 32  # Exemple de coordonnées et de taille pour le recadrage
image_with_rectangle=cv2.rectangle(image.copy(),(y,x),(y+h,x+w),(0,0,255),3)
plt.imshow(cv2.cvtColor(image_with_rectangle, cv2.COLOR_BGR2RGB))
```

```python
# Afficher l'image recadrée
cropped_image = image[y:y + h, x:x + w]

plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.title("Image recadrée")
plt.axis("off")
plt.show()
```

```python
print(cropped_image[:, :,0]),print(cropped_image[:, :,1]),print(cropped_image[:, :,2])
```

```python
# Séparation des canaux de couleur
b, g, r = cv2.split(image)
```

```python
# Affichage des canaux de couleur
fig, axes = plt.subplots(1, 3, figsize=(10, 4))
axes[0].imshow(b, cmap="Blues")
axes[0].set_title("Canal bleu")
axes[0].axis("off")

axes[1].imshow(g, cmap="Greens")
axes[1].set_title("Canal vert")
axes[1].axis("off")

axes[2].imshow(r, cmap="Reds")
axes[2].set_title("Canal rouge")
axes[2].axis("off")

plt.show()
```

```python
# Création d'un effet de couleur
blue_image = cv2.merge((b, np.zeros_like(g), np.zeros_like(r)))
```

```python
# Afficher l'image bleue
plt.imshow(cv2.cvtColor(blue_image, cv2.COLOR_BGR2RGB))
plt.title("Effet bleu")
plt.axis("off")
plt.show()
```
